```python
import pandas as pd
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 300
```

```python
def q_map(ca,cb,za=0,zb=0,):
    za_sqr = za**2-zb**2 + ca
    zb_sqr = 2*za*zb + cb

    return za_sqr, zb_sqr

def iter_q_map(ca,cb,n=1000,o='square'):
    za,zb = 0,0
    for i in range(n):
        # print(za,zb)
        za,zb = q_map(ca,cb,za,zb)
        r = za**2+zb**2
        if r>25:
            break
    if o=='square':
        return r
    elif o=='iter':
        return i
        
iter_q_map(0.1,0)
```

```python
def plot_m_set(x_start=-10,x_end=10,y_start=-10,y_end=10,ppu=10,step_ticks = None,plot_val='square'):
    Nx = (x_end-x_start)*ppu+1
    Ny = (y_end-y_start)*ppu+1
    Nx = int(Nx)
    Ny = int(Ny)
    data = np.zeros((Ny,Nx))
    # print(data.shape)
    l_x = np.arange(x_start,x_end+1/ppu,1/ppu)
    l_x = np.round(l_x,decimals=5,)
    l_y = np.arange(y_start,y_end+1/ppu,1/ppu)
    l_y = np.round(l_y,decimals=5,)

    
    nx=0
    for x in l_x:
        ny=0
        for y in l_y:
            # print(x,y)
            # data[nx,ny] = nx+ny
            data[ny,nx] = iter_q_map(x,y,o=plot_val)
            # if nx<100:
            #     if ny<100:
            #         data[nx,ny] = 500
            ny+=1
        nx+=1
        # break


    if step_ticks==None:
        dT_x = (x_end-x_start)/10
        dT_x = np.round(dT_x,decimals=5,)
        dT_y = (y_end-y_start)/10
        dT_y = np.round(dT_y,decimals=5,)
    else:
        dT_x = step_ticks
        dT_y = step_ticks
        
    map_labels_x = np.arange(Nx)
    map_labels_x = {j:i for i,j in zip(map_labels_x,l_x)}  
    labels_x = np.arange(x_start,x_end+dT_x,dT_x)
    labels_x = np.round(labels_x,decimals=5,)
    # print(labels_x)
    ticks_x = []
    for i in labels_x:
        # print(i,map_labels_x[i])
        try:
            ticks_x.append(map_labels_x[i])
        except:
            ticks_x.append(np.nan)


    
    map_labels_y = np.arange(Ny)
    map_labels_y = {j:i for i,j in zip(map_labels_y,l_y)}  
    labels_y = np.arange(y_start,y_end+dT_y,dT_y)
    labels_y = np.round(labels_y,decimals=5,)
    # print(labels_y)
    ticks_y = []
    for i in labels_y:
        # print(i,map_labels_y[i])
        try:
            ticks_y.append(map_labels_y[i])
        except:
            ticks_y.append(np.nan)
    
    fig, ax = plt.subplots()
    im = ax.imshow(
        data,
        origin='lower',
        interpolation='antialiased',
    )
    ax.set_xticks(
        ticks=ticks_x,
        labels=labels_x,
        rotation='vertical',
    )
    ax.set_yticks(
        ticks=ticks_y,
        labels=labels_y,
        rotation='horizontal',
    )
    ax.set_title("m set")
    fig.colorbar(im)
    plt.show()

            
plot_m_set(
    x_start=-1.7,
    x_end=0.7,
    y_start=-1.5,
    y_end=1.5,
    ppu=200,
    # step_ticks=0.1,
    plot_val='iter',
)
# plot_m_set()
```
